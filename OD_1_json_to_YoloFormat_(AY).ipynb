Class 선언하기

```python
class_names_dic = {'ConcreteCrack': 0,
                   'Efflorescene': 1,
                   'Exposure': 2,
                   'PaintDamage': 3,
                   'Spalling': 4,
                   'SteelDefect': 5}
# enumertate
```

Json 내용 중 내가 필요한 것(class, x, y, w, h)만 골라서 리스트에 담는 작업

```python
converted_label = [] 
for i in range(len(label_path_list)): 
    with open(label_path + '/' + label_path_list[i], encoding='UTF-8') as json_file: #i번째 파일을 열어 그걸 Json_file로 별명붙여줄게
        json_data = json.load(json_file) #열은 파일 내용 읽어와

        tmp_box_list = [] 
        for j in range(len((json_data['annotations']))):
            class_id = json_data['annotations'][j]['attributes']['class']
            width = float(json_data['images'][0]['width']) # image_width
            height = float(json_data['images'][0]['height']) # image_height
            bbox = json_data['annotations'][j]['bbox'] #bbox [center_x, center_y, w, h]
            
            #normalize 수행, 레이블들을 0~1사이의 값을 갖도록 조정해주기 위해 width, height로 나눠준다.
            center_x = bbox[0] / width
            center_y = bbox[1] / height
            w = bbox[2] / width
            h = bbox[3] / height
            
            temp = [class_names_dic[class_id], center_x, center_y, w, h] 
            tmp_box_list.append(temp) 
    
    converted_label.append(tmp_box_list)
```

```python
last_path = '/content/drive/MyDrive/Detection_Test/txt/'
for i in range(len(converted_label)):
    f = open(last_path + label_path_list[i].replace('.json', '.txt'), 'w')
    for j in range(len(converted_label[i])):
        f.write(str(converted_label[i][j][0])+' ')
        f.write(str(converted_label[i][j][1])+' ')
        f.write(str(converted_label[i][j][2])+' ')
        f.write(str(converted_label[i][j][3])+' ')
        f.write(str(converted_label[i][j][4])+'\n')
    f.close()
```

```python
temp = []
for i in range(len(label_path_list)):
  if '.json' in label_path_list[i]:
    a = label_path_list[i].replace('.json','')
    temp.append(a)
```
