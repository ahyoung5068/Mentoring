YOLO style Label로 변환하기 위한 함수 선언

```python
def convert(width, height, box):
    dw = 1./(width)
    dh = 1./(height)
    x = (float(box[0]) + float(box[1]))/2.0 - 1
    y = (float(box[2]) + float(box[3]))/2.0 - 1
    w = float(box[1]) - float(box[0])
    h = float(box[3]) - float(box[2])
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```
