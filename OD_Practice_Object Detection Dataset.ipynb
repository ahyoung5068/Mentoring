```python
# 각 이미지에 해당하는 file_name이랑, 이에 해당하는 bbox 좌표들, height, width 출력

d = dict()
for i in ls :
  name = i['images'][0]['file_name']
  content = dict()

  for j in range(len(i['annotations'])) :
    bname = "bbox" + str(j+1)
    bbox = i['annotations'][j]['bbox']
    bbox = [(bbox[0]+bbox[2])/2.0,(bbox[1]+bbox[3])/2.0]
    content[bname] = bbox
    
  content['width'] = i['images'][0]['width']
  content['height'] = i['images'][0]['height']


  d[name] = content
```

```python
import os
import json

path = '/content/drive/MyDrive/Detection Test/라벨링데이터/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.json')] 

data = []
for i in file_list_py:
    with open(path + i,'r', encoding='UTF8') as f :
        d = json.load(f)
        data.append(d)
#  Cyclic processing
d = dict()

for t in data:
    tmp = t['images'][0]['file_name'].split('.')[0]
    #filename = out_dir + tmp[0] + '.txt'

    content = dict()

    for j in range(len(t['annotations'])) :
      bname = "bbox" + str(j+1)
      bbox = t['annotations'][j]['bbox']
      bbox = [(bbox[0]+bbox[2])/2.0,(bbox[1]+bbox[3])/2.0]
      content[bname] = bbox
      
    content['width'] = t['images'][0]['width']
    content['height'] = t['images'][0]['height']

d[tmp] = content

```
